# How many owners have pets where the first letter of their name (OwnerName) matches the first letter of their pet's name (PetName)? E.g. Cookie and Charles.

```python
dx
```

```python
dq = dx[['OwnerName', 'PetName']]; dq
```

```python
dq['OwnerName_1'] = dq['OwnerName'].str[0]
dq['PetName_1'] = dq['PetName'].str[0]
```

```python
res = []

for _, i in dq.iterrows():
    if i['OwnerName_1'] == i['PetName_1']:
        res.append(i['OwnerName'])
```

```python
from collections import Counter
res = Counter(res)
total_count = sum(res.values())
print(f'Number of owners with their first letters of their names matching with that of their pets = {total_count}')
```

# What percentage of cities have more male pets than female pets?

```python
dw = dx[['City', 'Gender']]; dw
```

```python
male_counts = dw[dw['Gender'] == 'male']['City'].value_counts()
```

```python
male_counts = pd.DataFrame(male_counts, columns=['City']).reset_index()
```

```python
male_counts = male_counts.rename(columns = {'index': 'City', 'City': 'MaleCounts'})
```

```python
female_counts = dw[dw['Gender'] == 'female']['City'].value_counts()
```

```python
female_counts = pd.DataFrame(female_counts, columns=['City']).reset_index()
```

```python
female_counts = female_counts.rename(columns = {'index': 'City', 'City': 'FemaleCounts'})
```

```python
dg = male_counts.join(female_counts['FemaleCounts'])
```

```python
city = []
for _, i in dg.iterrows():
    if i['MaleCounts'] > i['FemaleCounts']:
        city.append(i['City'])
city
```

```python
num_city = len(city); num_city
```

```python
res = (num_city * 100) /dw['City'].nunique(); res
```

```python
print(f'{round(res)}% of cities have more male pets than female pets')
```
